```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path ="/content/drive/MyDrive/datasets/Employee.csv"
df = pd.read_csv(path)
df.head()
```

#*line*

```python
# @title JoiningYear

df['JoiningYear'].plot(kind='line', figsize=(8, 4), title='JoiningYear')
plt.show()
```

```python
# @title Age vs ExperienceInCurrentDomain


df.plot(kind='scatter', x='Age', y='ExperienceInCurrentDomain', s=32, alpha=.8)
```

```python
# @title JoiningYear vs PaymentTier

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='JoiningYear', y='PaymentTier', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Education

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Education').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title ExperienceInCurrentDomain

from matplotlib import pyplot as plt
df['ExperienceInCurrentDomain'].plot(kind='hist', bins=20, title='ExperienceInCurrentDomain')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Age

from matplotlib import pyplot as plt
df['Age'].plot(kind='hist', bins=20, title='Age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title PaymentTier

from matplotlib import pyplot as plt
df['PaymentTier'].plot(kind='hist', bins=20, title='PaymentTier')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title JoiningYear

from matplotlib import pyplot as plt
df['JoiningYear'].plot(kind='hist', bins=20, title='JoiningYear')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
