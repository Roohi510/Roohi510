* Country name: This column contains the names of different countries represented in the dataset.
* Regional indicator: Indicates the region or geographical area to which each country belongs.
* Ladder score: Represents the happiness score or subjective well-being of individuals in each country, often measured through surveys or polls.
* Standard error of ladder score: Provides information about the variability or uncertainty associated with the ladder scores.
* Upper whisker: Indicates the upper limit of the confidence interval for the ladder score.
* Lower whisker: Indicates the lower limit of the confidence interval for the ladder score.
* Logged GDP per capita: Represents the GDP per capita of each country after being logarithmically transformed.
* Social support: Refers to the perceived social support available to individuals in each country.
* Healthy life expectancy: Indicates the average life expectancy and overall health status of individuals in each country.
*Freedom to make life choices: Measures the degree of freedom individuals perceive in making life choices, such as career, education, and personal relationships.
* Generosity: Represents the level of generosity or altruism observed within each country's population.
* Perceptions of corruption: Measures the perceived level of corruption within each country's institutions and society.
* Ladder score in Dystopia: Represents the hypothetical happiness score in a dystopian scenario where all factors are at their lowest possible levels.
* Explained by: Log GDP per capita, Social support, Healthy life expectancy, Freedom to make life choices, Generosity, Perceptions of corruption: These columns provide insights into how much each factor contributes to the overall ladder score, based on statistical analysis.
* Dystopia + residual: Represents the sum of the hypothetical Dystopia ladder score and any residual factors not explained by the other variables.

```python
from scipy.stats import ttest_ind, wilcoxon, ttest_rel
def hypo_test(reg1,reg2):
    r1 = df[df['Regional indicator']==reg1]['Ladder score']
    r2 = df[df['Regional indicator']==reg2]['Ladder score']
    t1="Hypotheis test between regions: {} vs {}".format(reg1,reg2)
    stats, p_value= ttest_ind(r1, r2)

    if p_value <0.05:
        t2="since p_value : {} is < 0.05, we can reject the Null Hypothesis and say there are significant difference".format(np.round(p_value,3))
    else:
        t2="P_value: {} is >0.05, we can't reject the Null Hypothesis, so we say there is no difference".format(np.round(p_value,3))
    fig=plt.figure(figsize=(15,3))
    plt.title(t1,family='Serif', weight='bold', size=20)
    plt.text(x=plt.xlim()[1]/2,y=plt.ylim()[1]/2,s=t2,ha='center', family='Serif', weight='bold', size=15)
    for pos in ['right','top','bottom','left']:
        plt.gca().spines[pos].set_visible(False)
    plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
    plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)
    return
```

```python
reg=df['Regional indicator'].unique()
hypo_test(reg[0],reg[1])
hypo_test(reg[0],reg[-1])
```
