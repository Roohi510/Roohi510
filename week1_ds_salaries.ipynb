```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
import random
classes=["ai","ml","py","iot","ds"]

class1_students =np.random.randint(15,45,(5))
class2_students =np.random.randint(15,45,(5))
class3_students =np.random.randint(15,45,(5))
print(class1_students)
print(class2_students)
print(class3_students)
plt.title("no of students in each class")
plt.xlabel("subjects")
plt.ylabel("no. of students")
plt.bar(classes,class1_students,color="m")
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['work_year']
  ys = series['salary']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = _df_8.sort_values('work_year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('employment_type')):
  _plot_series(series, series_name, i)
  fig.legend(title='employment_type', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('work_year')
_ = plt.ylabel('salary')
```
