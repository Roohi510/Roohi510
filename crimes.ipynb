```python
primary_type_counts = df['Primary_Type'].value_counts()

# Calculate the total count
total_count = primary_type_counts.sum()

# Threshold for "Other" category
threshold = total_count * 0.005

# Filter primary types with contribution less than the threshold
small_contributions = primary_type_counts[primary_type_counts < threshold]

# Sum up small contributions and assign them to the 'Other' category
other_count = small_contributions.sum()

# Update primary type counts
primary_type_counts = primary_type_counts[primary_type_counts >= threshold]
primary_type_counts['Other'] = other_count

# Plotting the pie chart
plt.figure(figsize=(16, 9))
plt.pie(primary_type_counts, labels=primary_type_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Primary Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
